# Import of Libraries

```python
import pandas as pd
pd.__version__
```

# **Data upload**

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations from stooq.pl
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

apple = fetch_financial_data('AAPL')
amazone = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
```

```python
uber.info()
```

```python
uber.head()
```

# **Preprocessing**

```python
apple.columns = ['apple_' + col.lower() for col in apple.columns]
amazone.columns = ['amazone_' + col.lower() for col in amazone.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
```

```python
uber.head()
```

# Data concatenation

```python
df = pd.concat(objs=[apple, amazone, google, uber], axis=1)
df
```

```python
df.describe().T
```

```python
#to improve the data format
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
df.describe().T
```

```python
df.corr()
```

```python
#to extract data from closed sessions
closes = [col for col in df.columns if col.endswith('close')]
closes
```

```python
df[closes].corr()
```
