# Import of Libraries

```python
import numpy as np
import pandas as pd
```

# Data upload

```python
def fetch_financial_data(company='META'):
    """
    This function fetches stock market quotations from stooq.pl
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df.head()
```

# Saving/opening CSV files

```python
df.to_csv('meta.csv')
```

```python
# July sessions
df_jul = df[(df.index.month == 7) & (df.index.year == 2022)]
df_jul
```

```python
df_jul.to_csv('meta_jul.csv')
```

```python
new_df = pd.read_csv('meta_jul.csv')
new_df
```

# Saving/opening XLSX files

```python
df_jul.to_excel('meta_jul.xlsx')
```

```python
new_df = pd.read_excel('meta_jul.xlsx')
new_df
#date is loaded as a column, not as an index
```

```python
#to set a date as index value
new_df = pd.read_excel('meta_jul.xlsx', index_col=0)
new_df
```
