```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('cat.jpg')
img
```

```python
img.shape
```

```python
plt.imshow(img)
_ = plt.axis('off')
```

```python
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('cat.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('dog.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('hamster.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('parrots.jpg'))
```

```python
# the same via for_loop
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['cat.jpg', 'dog.jpg', 'hamster.jpg', 'parrots.jpg']):
  plt.subplot(221 + idx)
  plt.imshow(mpimg.imread(img))
```

```python
# to make 1 x 4 multiplot

fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['cat.jpg', 'dog.jpg', 'hamster.jpg', 'parrots.jpg']):
  plt.subplot(141 + idx)
  plt.imshow(mpimg.imread(img))
```
