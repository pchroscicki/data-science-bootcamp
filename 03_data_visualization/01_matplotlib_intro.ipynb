# MATPLOTLIB

Website: https://matplotlib.org/

```python
# import matplotlib
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

# Line chart

```python
plt.plot([0, 1], [0, 3])
```

```python
#solid curve
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Chart1')
_ = plt.legend()
```

```python
#broken curve
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Chart1')
_ = plt.legend()
```

```python
# sinusoid
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)

plt.plot(x, x_sin)
```

```python
# sinusoid, two lines at one plot
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

# Multiple plots

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [8, 12, 0])
```

```python
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 2, 5], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 2, 5], 'ro')
```

# Bar charts

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 2.0, 6.2])
# vertical bar charts are plotted by default
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 2.0, 6.2])
```

```python
# horizontal bar charts
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 2.0, 6.2])
```

# Scatter charts

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
plt.scatter(x=x1, y=x2)
```

```python
# to modify the dot size
x3 = np.random.randn(300) * 50
plt.scatter(x=x1, y=x2, s=x3)
```

```python
# alpha - sets the transparency of dots
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
